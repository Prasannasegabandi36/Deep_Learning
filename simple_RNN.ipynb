```python
import numoy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
## load the imbd dataset
max_features=10000 ## vocabulary size
(X_train,Y_train),(X_test,Y_test)=imdb.load_data(num_words=max_features)
## print the shape of data
print(f'Training data shape:{X_train.shape},Training labels shape:{Y_train.shape}')
print(f'Training data shape:{X_train.shape},Testing labels shape:{Y_train.shape}')
```

```python
X_train[0],Y_train[0]
```

```python
##inspect a simple review and its label
sample_review=X_train[0]
sample_label=Y_train[0]
print(f"sample review(as integers):{sample_review}")
print(f"sample label:{sample_label}")
```

```python
##mapping of words index bacl to words(for understanding)
word_index=imdb.get_word_index()
word_index
reverse_word_index={value: key for key,value in word_index.items()}
reverse_word_index
```

```python
decoded_review=' '.join([reverse_word_index.get(i - 3 ,'?')for i in sample_review])
decoded_review
```

```python
from tensorflow.keras.preprocessing import sequence
max_len=500
X_train=sequence.pad_sequences(X_train,max_len=max_len)
X_test=sequence.pad_sequences(X_test,max_len=max_len)
X_train
```

```python
X_train[0]
```

```python
# train simple rnn
model=Sequential()
model.add(Embedding(max_features,128,input_length==max_len)) # embedding layer
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.complie(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##create an instance of earlystopping callback
from tensorflow.keras.callbacks import EarlyStopping
EarlyStopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
EarlyStopping
```

```python
## TRain the model with early stopping
history=model.fit(
    X_train,Y_train,epochs=10,batch_size=32,
    validation_split=0.2,
    callback=[EarlyStopping]
)
```

```python
## save model file
model.save('simple_rnn_imdb.h5')
```
