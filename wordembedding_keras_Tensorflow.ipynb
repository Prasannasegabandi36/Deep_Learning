```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentnces
sent=['The cup of tea is hot',

'The bottle of water is full',

'The bowl of rice is on the table',

'The plate of fruits is fresh',

'The jar of honey is sweet',

'The slice of bread is warm',

'The glass of juice is cold',

'The packet of sugar is open'.]
```

```python
sent
```

```python
#define the vocabulary size
voc_size=10000
```

```python
## one hot representation
one_hot_repr=[one_hot(words,voc_size)for word in sent]
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
#from tensorflow.keras.precoessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy  as np
```

```python
#must all sent are same length or size thats y we use pad
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#feature represntation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.complie('adam','mse')
```
